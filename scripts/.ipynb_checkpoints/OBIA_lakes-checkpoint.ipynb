## Glacial Lake Mapping in Southeast Iceland using Sentinel-2 via OpenEO

Replicates eCognition Multiresolution Segmentation (MRS) approach based on [Dabiri et al 2021](moz-extension://bc18e54e-46b5-47f8-9750-d48b8835d745/enhanced-reader.html?openApp&pdf=https%3A%2F%2Faustriaca.at%2F0xc1aa5576%25200x003c9b50.pdf) in Python.
 
Workflow per year (2016–2025):
  1. Load Sentinel-2 L2A summer composite (July-June) via openEO
  2. Compute NDVI, NDWI, NDSI, and Brightness
  3. Segment image using scikit-image SLIC (analogous to MRS)
  4. Classify water segments using NDWI >= 0.3, refined by NDVI and size
  5. Export polygon outlines per year as GeoPackage / GeoJSON

```python
# Imports:

import os
import json
import numpy as np
import rasterio
from rasterio.transform import from_bounds
from rasterio.crs import CRS
from rasterio.features import shapes as rio_shapes
import geopandas as gpd
from shapely.geometry import shape, mapping
from skimage.segmentation import slic, mark_boundaries
from skimage.measure import regionprops
import openeo
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
# Bounding box: Southeast Iceland (Vatnajökull glacier lake area)
# Adjust as needed for your specific study area
BBOX = {
    "west":  -16.5,
    "south":  63.97,
    "east":  -16.1,
    "north":  64.2,
}
BBOX_EPSG = 4326

# Reproject to UTM zone 27N (EPSG:32627) for metric distance calculations
TARGET_CRS = "EPSG:32627"
 
# Temporal range: summer months (July–September) per year
YEARS = list(range(2016, 2026))  # 2016–2025 inclusive
SUMMER_START_MM_DD = "06-01"
SUMMER_END_MM_DD   = "09-30"
 
# Cloud masking: max acceptable cloud cover per scene
MAX_CLOUD_COVER = 20 # percent
 
# Output directory
OUTPUT_DIR = "../data/output/lake_polygons"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

### Segmentation parameters

```python
SEG_PARAMS = {
    "n_segments":    800,   # analogous to Scale Parameter (SP) 800
    "compactness":   0.4,   # eCognition compactness 0.4
    "sigma":         0.8,
    "min_size_px":   15,    # minimum segment size in pixels
}
```

```python
# Classification thresholds
NDWI_WATER_THRESHOLD = 0.3   # NDWI >= 0.3 → water
NDVI_VEGE_THRESHOLD  = 0.1   # NDVI < 0.1 to exclude vegetation patches
MIN_LAKE_AREA_M2     = 10000  # 1 ha minimum lake size (adjust as needed)
MAX_LAKE_AREA_M2     = 30000000    # upper bound to exclude sea/ocean polygons
```

### OpenEO connection

```python
def connect_openeo(backend_url: str = "https://openeo.dataspace.copernicus.eu") -> openeo.Connection:
    """
    Connect and authenticate to an OpenEO backend.
    Uses Copernicus Dataspace by default (free registration at dataspace.copernicus.eu).
    Alternatively use: https://openeo.cloud
    """
    print(f"Connecting to OpenEO backend: {backend_url}")
    conn = openeo.connect(backend_url)
    conn.authenticate_oidc()  # Opens browser for OIDC login on first run
    print("✓ Connected and authenticated.")
    return conn
```

### Band loading and Index calculation

```python
def build_openeo_job(conn: openeo.Connection, year: int) -> openeo.rest.datacube.DataCube:
    """
    Build an OpenEO process graph that:
      - Loads Sentinel-2 L2A for summer of the given year
      - Applies cloud masking via SCL (Scene Classification Layer)
      - Reduces to a median composite
      - Returns bands: B02, B03, B04, B08, B11, B12
    """
    temporal_extent = [
        f"{year}-{SUMMER_START_MM_DD}",
        f"{year}-{SUMMER_END_MM_DD}",
    ]
 
    # Sentinel-2 L2A band mapping:
    # B02=Blue, B03=Green, B04=Red, B08=NIR, B11=SWIR1, B12=SWIR2
    # SCL = Scene Classification Layer (for cloud masking)
    bands = ["B02", "B03", "B04", "B08", "B11", "B12", "SCL"]
 
    dc = conn.load_collection(
        "SENTINEL2_L2A",
        spatial_extent=BBOX,
        temporal_extent=temporal_extent,
        bands=bands,
        max_cloud_cover=MAX_CLOUD_COVER,
    )
 
    # Cloud masking: keep only pixels classified as
    # vegetation(4), bare soil(5), water(6), unclassified(7), snow/ice(11)
    # Exclude cloud shadow(3), cloud medium(8), cloud high(9), thin cirrus(10)
    scl = dc.band("SCL")
    cloud_mask = (
        (scl == 3) | (scl == 8) | (scl == 9) | (scl == 10)
    )
    dc = dc.mask(cloud_mask)
 
    # Reduce time dimension: median composite
    dc = dc.filter_bands(["B02", "B03", "B04", "B08", "B11", "B12"])
    dc = dc.reduce_dimension(dimension="t", reducer="median")
 
    return dc
 
 
def download_composite(conn: openeo.Connection, year: int, out_path: str) -> str:
    """
    Download the median summer composite for a given year as a GeoTIFF.
    Skips download if file already exists.

    Uses an isolated temporary subdirectory per year so that OpenEO's
    generic output filenames (e.g. openEO_0.tif) never collide across
    years, and the correct file is always moved to `out_path`.
    """
    if os.path.exists(out_path):
        print(f"  ↳ Using cached composite: {out_path}")
        return out_path
 
    print(f"  ↳ Building OpenEO job for {year}...")
    dc = build_openeo_job(conn, year)

    # Download into a year-specific temp folder to avoid cross-year collisions
    tmp_dir = os.path.join(os.path.dirname(out_path), f"_tmp_{year}")
    os.makedirs(tmp_dir, exist_ok=True)
 
    # Save result as GeoTIFF
    result = dc.save_result(format="GTiff")
    job = result.create_job(title=f"glacial_lakes_SE_Iceland_{year}")
    print(f"  ↳ Submitting batch job (ID: {job.job_id})...")
    job.start_and_wait()
 
    job.get_results().download_files(target=tmp_dir)
 
    # Find all TIFFs downloaded into the temp folder
    downloaded_tifs = sorted([
        f for f in os.listdir(tmp_dir) if f.lower().endswith(".tif")
    ])

    if not downloaded_tifs:
        raise FileNotFoundError(
            f"No .tif files found in {tmp_dir} after downloading year {year}. "
            f"Contents: {os.listdir(tmp_dir)}"
        )

    if len(downloaded_tifs) > 1:
        print(f"  ⚠ Multiple TIFs found for {year}: {downloaded_tifs} — using first.")

    src_file = os.path.join(tmp_dir, downloaded_tifs[0])
    os.rename(src_file, out_path)

    # Clean up temp dir (should now be empty)
    try:
        os.rmdir(tmp_dir)
    except OSError:
        pass  # non-empty (extra files) — leave it

    print(f"  ✓ Downloaded: {out_path}")
    return out_path
```

```python
def compute_indices(blue, green, red, nir, swir1):
    """
    Compute spectral indices from surface reflectance bands (0–1 scale).
 
    Returns: ndvi, ndwi, ndsi, brightness (all float32 arrays)
    """
    eps = 1e-10  # avoid divide-by-zero
 
    # NDVI = (NIR - Red) / (NIR + Red)
    ndvi = (nir - red) / (nir + red + eps)
 
    # NDWI (Gao) = (NIR - SWIR1) / (NIR + SWIR1)  → open water
    # Note: some studies use McFeeters NDWI = (Green - NIR)/(Green + NIR)
    # Here we use McFeeters as it's more commonly used for lake delineation
    ndwi = (green - nir) / (green + nir + eps)
 
    # NDSI = (Green - SWIR1) / (Green + SWIR1)  → snow/ice discrimination
    ndsi = (green - swir1) / (green + swir1 + eps)
 
    # Brightness = mean of visible bands (Blue, Green, Red)
    brightness = (blue + green + red) / 3.0
 
    # Clip to valid index range
    ndvi = np.clip(ndvi, -1, 1)
    ndwi = np.clip(ndwi, -1, 1)
    ndsi = np.clip(ndsi, -1, 1)
 
    return ndvi.astype(np.float32), ndwi.astype(np.float32), \
           ndsi.astype(np.float32), brightness.astype(np.float32)
```

### Segmentation

```python
def segment_image(blue, green, red, nir, ndvi, ndwi, ndsi, brightness,
                  seg_params: dict) -> np.ndarray:
    """
    Segment the image using SLIC superpixels.
 
    SLIC is the closest open-source analogue to eCognition MRS:
      - n_segments ↔ Scale Parameter (SP)
      - compactness ↔ shape/compactness weighting
      - Uses both spectral bands and derived indices as features
 
    The feature stack mirrors the eCognition input:
      spectral bands + NDVI + NDWI + NDSI + brightness
    """
    # Stack feature layers: bands + indices
    # Normalise each layer to [0, 1] for balanced SLIC weighting
    def norm(arr):
        mn, mx = np.nanpercentile(arr, 2), np.nanpercentile(arr, 98)
        if mx == mn:
            return np.zeros_like(arr, dtype=np.float32)
        return np.clip((arr - mn) / (mx - mn), 0, 1).astype(np.float32)
 
    feature_stack = np.dstack([
        norm(blue),
        norm(green),
        norm(red),
        norm(nir),
        norm(ndvi),
        norm(ndwi),
        norm(ndsi),
        norm(brightness),
    ])
 
    # Replace NaN with 0 (masked/cloudy pixels)
    feature_stack = np.nan_to_num(feature_stack, nan=0.0)
 
    segments = slic(
        feature_stack,
        n_segments=seg_params["n_segments"],
        compactness=seg_params["compactness"],
        sigma=seg_params["sigma"],
        start_label=1,
        channel_axis=-1,
    )
 
    return segments
```

### Classification

```python
def classify_water_segments(segments: np.ndarray,
                             ndwi: np.ndarray,
                             ndvi: np.ndarray,
                             seg_params: dict) -> np.ndarray:
    """
    Classify segments as water using object-based rules:
 
    Primary:   mean NDWI >= 0.3
    Refinement:
      - mean NDVI < 0.1  (exclude vegetation)
      - size >= min_size_px (exclude tiny noise segments)
 
    Returns a binary mask (1=water, 0=non-water).
    """
    water_mask = np.zeros_like(segments, dtype=np.uint8)
 
    props = regionprops(segments)
    for prop in props:
        seg_id = prop.label
        px_mask = segments == seg_id
 
        mean_ndwi = float(np.nanmean(ndwi[px_mask]))
        mean_ndvi = float(np.nanmean(ndvi[px_mask]))
        size_px   = int(np.sum(px_mask))
 
        # Primary criterion
        if mean_ndwi < NDWI_WATER_THRESHOLD:
            continue
 
        # Refinement: exclude vegetation
        if mean_ndvi >= NDVI_VEGE_THRESHOLD:
            continue
 
        # Refinement: exclude tiny artefacts
        if size_px < seg_params["min_size_px"]:
            continue
 
        water_mask[px_mask] = 1
 
    return water_mask
```

### Vectorisation

```python
def vectorise_water_mask(water_mask: np.ndarray,
                          transform: rasterio.transform.Affine,
                          src_crs: CRS,
                          year: int) -> gpd.GeoDataFrame:
    """
    Convert the binary water mask to polygon geometries.
    Filters by area thresholds. Reprojects to TARGET_CRS.
    """
    polys = []
    for geom, val in rio_shapes(water_mask, mask=water_mask.astype(np.uint8),
                                  transform=transform):
        if val == 1:
            polys.append(shape(geom))
 
    if not polys:
        print(f"  ⚠ No water polygons found for {year}")
        return gpd.GeoDataFrame(columns=["geometry", "year", "area_m2"],
                                crs=src_crs)
 
    gdf = gpd.GeoDataFrame({"geometry": polys}, crs=src_crs)
 
    # Reproject to metric CRS for area calculation
    gdf = gdf.to_crs(TARGET_CRS)
    gdf["area_m2"] = gdf.geometry.area
    gdf["year"] = year
 
    # Filter by area
    gdf = gdf[(gdf["area_m2"] >= MIN_LAKE_AREA_M2) &
              (gdf["area_m2"] <= MAX_LAKE_AREA_M2)].copy()
    gdf = gdf.reset_index(drop=True)
 
    print(f"  ✓ {len(gdf)} lake polygons detected for {year} "
          f"(total area: {gdf['area_m2'].sum()/1e6:.2f} km²)")
    return gdf
```

### Main pipeline

```python
def process_year(conn: openeo.Connection, year: int) -> gpd.GeoDataFrame:
    """
    Full pipeline for a single year:
      download → compute indices → segment → classify → vectorise
    """
    print(f"\n{'='*55}")
    print(f"  Processing year: {year}")
    print(f"{'='*55}")
 
    # 1. Download composite
    tif_path = os.path.join(OUTPUT_DIR, f"composite_{year}.tif")
    download_composite(conn, year, tif_path)
 
    # 2. Load bands from GeoTIFF
    with rasterio.open(tif_path) as src:
        # Band order from OpenEO: B02, B03, B04, B08, B11, B12
        # (1-indexed in rasterio)
        blue   = src.read(1).astype(np.float32) / 10000.0  # DN → reflectance
        green  = src.read(2).astype(np.float32) / 10000.0
        red    = src.read(3).astype(np.float32) / 10000.0
        nir    = src.read(4).astype(np.float32) / 10000.0
        swir1  = src.read(5).astype(np.float32) / 10000.0
        # swir2  = src.read(6).astype(np.float32) / 10000.0  # available if needed
 
        transform = src.transform
        src_crs   = src.crs
 
        # Mask no-data (0 values in L2A)
        nodata_mask = (blue == 0) | (nir == 0)
        blue  = np.where(nodata_mask, np.nan, blue)
        green = np.where(nodata_mask, np.nan, green)
        red   = np.where(nodata_mask, np.nan, red)
        nir   = np.where(nodata_mask, np.nan, nir)
        swir1 = np.where(nodata_mask, np.nan, swir1)
 
    print(f"  ↳ Image shape: {blue.shape}")
 
    # 3. Compute indices
    print("  ↳ Computing spectral indices...")
    ndvi, ndwi, ndsi, brightness = compute_indices(blue, green, red, nir, swir1)
 
    # 4. Select segmentation parameters
    seg_params = SEG_PARAMS
    print(f"  ↳ Segmenting with SP={seg_params['n_segments']}, "
          f"compactness={seg_params['compactness']}...")
    segments = segment_image(blue, green, red, nir,
                              ndvi, ndwi, ndsi, brightness,
                              seg_params)
 
    # 5. Classify water segments
    print("  ↳ Classifying water segments (NDWI >= 0.3)...")
    water_mask = classify_water_segments(segments, ndwi, ndvi, seg_params)
 
    # 6. Vectorise
    print("  ↳ Vectorising water mask...")
    gdf = vectorise_water_mask(water_mask, transform, src_crs, year)
 
    return gdf
 
 
def main():
    # Connect to OpenEO
    conn = connect_openeo()
 
    all_years_gdfs = []
 
    for year in tqdm(YEARS, desc="Processing years"):
        try:
            gdf = process_year(conn, year)
            if len(gdf) > 0:
                # Save per-year GeoPackage
                out_gpkg = os.path.join(OUTPUT_DIR, f"glacial_lakes_{year}.gpkg")
                gdf.to_file(out_gpkg, driver="GPKG")
                print(f"  ✓ Saved: {out_gpkg}")
 
                all_years_gdfs.append(gdf)
        except Exception as e:
            print(f"  ✗ Error processing {year}: {e}")
            continue
 
    # Combine all years into a single multi-year GeoPackage
    if all_years_gdfs:
        combined = gpd.pd.concat(all_years_gdfs, ignore_index=True)
        combined_path = os.path.join(OUTPUT_DIR, "glacial_lakes_2016_2025_all.gpkg")
        combined.to_file(combined_path, driver="GPKG")
        print(f"\n✓ Combined dataset saved: {combined_path}")
        print(f"  Total polygons across all years: {len(combined)}")
 
        # Also export as GeoJSON for easy inspection
        combined_json = os.path.join(OUTPUT_DIR, "glacial_lakes_2016_2025_all.geojson")
        combined.to_file(combined_json, driver="GeoJSON")
        print(f"  GeoJSON also saved: {combined_json}")
 
 
if __name__ == "__main__":
    main()
```
